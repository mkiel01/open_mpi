```python
# mpiexec -machinefile ./1_node_with_8_process -np 2 ./ex_2_fin_new_c 

import pandas as pd
import matplotlib.pyplot as plt

# Data from measurements
data = {
    "Total Message Size (B)": [128000, 1024000, 8192000, 65536000],
    "Throughput (Mbit/s)": [124.481787, 1526.889433, 14895.168346, 105385.984161],
    "Latency (ms)": [0.000001, 0.000000, 0.000000, 0.000000],
    "Message Size (B)": [8, 64, 512, 4096],  # Added missing comma here
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Plot throughput vs. total message size
plt.figure(figsize=(10, 5))
plt.plot(df["Total Message Size (B)"], df["Throughput (Mbit/s)"], marker='o')
plt.title('Throughput vs. Total Message Size - Standard Communication 1 node')
plt.xlabel('Total Message Size (B)')
plt.ylabel('Throughput (Mbit/s)')
plt.grid(True)
plt.show()

# Plot latency for short messages vs. total message size
plt.figure(figsize=(10, 5))
plt.plot(df["Message Size (B)"], df["Latency (ms)"] * 1000, marker='o')
plt.title('Latency for Short Messages vs. Message Size - Standard Communication 1 node')
plt.xlabel('Message Size (B)')
plt.ylabel('Latency (ms)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dane
approximations = [3.141652, 3.141547, 3.141652]
num_points = [61500000, 615000000, 4270000000]
ideal_pi = [3.141592653589793] * len(approximations)  # Idealne przybliżenie liczby pi

# Wykres
plt.figure(figsize=(10, 6))
plt.plot(num_points, approximations, marker='o', linestyle='-', color='b', label='Przybliżenie liczby pi')
plt.plot(num_points, ideal_pi, linestyle='--', color='r', label='Idealna liczba pi')
plt.xscale('log')  # Skala logarytmiczna dla liczby punktów
plt.xlabel('Liczba punktów')
plt.ylabel('Przybliżenie liczby pi')
plt.title('Dokładność przybliżenia liczby pi metodą Monte Carlo')
plt.legend()
plt.grid(True)
plt.show()
```
